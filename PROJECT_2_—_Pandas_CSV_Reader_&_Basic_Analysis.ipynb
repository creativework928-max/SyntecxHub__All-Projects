```python
import pandas as pd
import numpy as np
import io
print(f"Pandas version: {pd.__version__}")
```

## 2.1 — Create Sample Dataset & Read CSV/Excel

```python
# -------------------------------------------------------
# Create a sample CSV file to simulate reading from file
# -------------------------------------------------------
csv_data = """Name,Age,Department,Salary,Experience,JoinDate,City,Gender
Alice,29,Engineering,75000,5,2019-03-15,New York,Female
Bob,35,Marketing,62000,10,2014-07-22,Chicago,Male
Carol,42,Engineering,95000,18,2006-01-10,New York,Female
David,27,HR,45000,3,2021-05-05,Houston,Male
Eva,31,Engineering,82000,7,2017-09-30,New York,Female
Frank,38,Marketing,58000,12,2012-11-18,Chicago,Male
Grace,25,HR,42000,2,2022-02-14,Houston,Female
Henry,45,Engineering,110000,20,2004-06-01,San Francisco,Male
Iris,33,Marketing,67000,9,2015-08-25,Chicago,Female
Jack,29,HR,48000,4,2020-04-10,Houston,Male
Karen,36,Engineering,88000,12,2012-03-20,San Francisco,Female
Liam,41,Marketing,72000,15,2009-10-08,Chicago,Male
Mia,28,Engineering,71000,4,2020-07-17,New York,Female
Noah,34,HR,50000,8,2016-12-03,Houston,Male
Olivia,30,Marketing,65000,6,2018-06-28,Chicago,Female
"""

# Save to CSV file
with open('employees.csv', 'w') as f:
    f.write(csv_data)

# ---- Read CSV ----
df = pd.read_csv('employees.csv')
print("✅ CSV file read successfully!")
print(f"Shape: {df.shape} — {df.shape[0]} rows × {df.shape[1]} columns")
```

## 2.2 — Inspect DataFrame: head, tail, dtypes, info

```python
print("--- df.head() ---")
print(df.head())

print("\n--- df.tail(3) ---")
print(df.tail(3))

print("\n--- df.dtypes ---")
print(df.dtypes)

print("\n--- df.info() ---")
df.info()

print("\n--- df.shape ---")
print(df.shape)

print("\n--- df.columns ---")
print(df.columns.tolist())

print("\n--- df.index ---")
print(df.index)

print("\n--- df.sample(5) ---")
print(df.sample(5, random_state=42))
```

## 2.3 — Summary Statistics (mean, median, min, max, count)

```python
# --- Overall describe ---
print("--- df.describe() ---")
print(df.describe())

print("\n--- describe (include all) ---")
print(df.describe(include='all'))

# --- Individual statistics ---
numeric_cols = ['Age', 'Salary', 'Experience']

print("\n" + "="*50)
print("INDIVIDUAL STATISTICS")
print("="*50)

for col in numeric_cols:
    print(f"\n📊 {col}:")
    print(f"   Mean   : {df[col].mean():.2f}")
    print(f"   Median : {df[col].median():.2f}")
    print(f"   Mode   : {df[col].mode()[0]}")
    print(f"   Min    : {df[col].min()}")
    print(f"   Max    : {df[col].max()}")
    print(f"   Count  : {df[col].count()}")
    print(f"   Std    : {df[col].std():.2f}")
    print(f"   Var    : {df[col].var():.2f}")
    print(f"   25%    : {df[col].quantile(0.25)}")
    print(f"   75%    : {df[col].quantile(0.75)}")

# Group-wise stats
print("\n--- Salary by Department ---")
print(df.groupby('Department')['Salary'].agg(['mean','median','min','max','count']).round(2))

print("\n--- Salary by Gender ---")
print(df.groupby('Gender')['Salary'].agg(['mean','median','min','max']).round(2))

print("\n--- Value Counts: Department ---")
print(df['Department'].value_counts())

print("\n--- Value Counts: City ---")
print(df['City'].value_counts())
```

## 2.4 — Filter Rows, Select Columns & Slice Subsets

```python
# --- Select columns ---
print("--- Single Column ---")
print(df['Name'].head())

print("\n--- Multiple Columns ---")
print(df[['Name', 'Department', 'Salary']].head())

# --- Filter rows ---
print("\n--- Salary > 70,000 ---")
high_earners = df[df['Salary'] > 70000]
print(high_earners[['Name','Department','Salary']])

print("\n--- Engineering Department ---")
engineers = df[df['Department'] == 'Engineering']
print(engineers[['Name','Salary','Experience']])

print("\n--- Age between 30 and 40 ---")
mid_age = df[(df['Age'] >= 30) & (df['Age'] <= 40)]
print(mid_age[['Name','Age','Department']])

print("\n--- Engineering OR Salary > 80000 ---")
combined = df[(df['Department'] == 'Engineering') | (df['Salary'] > 80000)]
print(combined[['Name','Department','Salary']])

print("\n--- NOT from Chicago ---")
not_chicago = df[df['City'] != 'Chicago']
print(not_chicago[['Name','City']])

print("\n--- isin() filter (New York or Houston) ---")
ny_houston = df[df['City'].isin(['New York', 'Houston'])]
print(ny_houston[['Name','City','Salary']])

# --- loc and iloc ---
print("\n--- .loc (label-based) rows 0-4, specific columns ---")
print(df.loc[0:4, ['Name','Salary','Department']])

print("\n--- .iloc (position-based) first 3 rows, first 4 cols ---")
print(df.iloc[0:3, 0:4])

# --- Sort ---
print("\n--- Sorted by Salary (descending) ---")
print(df.sort_values('Salary', ascending=False)[['Name','Salary','Department']].head(5))

print("\n--- Sorted by Department then Salary ---")
print(df.sort_values(['Department','Salary'], ascending=[True, False])[['Name','Department','Salary']])
```

## 2.5 — Save Filtered Results to CSV & Excel

```python
# ---- Save full DataFrame to CSV ----
df.to_csv('employees_full.csv', index=False)
print("✅ Full data saved to employees_full.csv")

# ---- Save filtered results ----
high_earners = df[df['Salary'] > 70000]
high_earners.to_csv('high_earners.csv', index=False)
print(f"✅ High earners ({len(high_earners)} rows) saved to high_earners.csv")

engineers = df[df['Department'] == 'Engineering']
engineers.to_csv('engineers.csv', index=False)
print(f"✅ Engineers ({len(engineers)} rows) saved to engineers.csv")

# ---- Save to Excel (.xlsx) with multiple sheets ----
with pd.ExcelWriter('employees_report.xlsx', engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='All Employees', index=False)
    high_earners.to_excel(writer, sheet_name='High Earners', index=False)
    engineers.to_excel(writer, sheet_name='Engineers', index=False)

    # Summary sheet
    summary = df.groupby('Department')['Salary'].agg(['mean','min','max','count']).round(2)
    summary.columns = ['Avg Salary', 'Min Salary', 'Max Salary', 'Headcount']
    summary.to_excel(writer, sheet_name='Department Summary')

print("✅ Excel report saved to employees_report.xlsx (4 sheets)")

# ---- Verify: read back and check ----
df_back = pd.read_csv('employees_full.csv')
print(f"\nVerification — Read back CSV shape: {df_back.shape}")

xl_back = pd.read_excel('employees_report.xlsx', sheet_name='All Employees')
print(f"Verification — Read back Excel shape: {xl_back.shape}")

print("\n✅ Project 2 — Pandas CSV Reader & Basic Analysis Complete!")
```
